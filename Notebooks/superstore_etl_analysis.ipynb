## 4. Limpeza e Tratamento

```python
df["sales"] = (
    df["sales"]
    .astype(str)
    .str.replace(",", "", regex=False)
)

df["sales"] = pd.to_numeric(df["sales"])
```

```python
df["order_date"] = pd.to_datetime(
    df["order_date"],
    format="mixed",
    dayfirst=True
)

df["ship_date"] = pd.to_datetime(
    df["ship_date"],
    format="mixed",
    dayfirst=True
)
```

## 5. Engenharia de Atributos

```python
df["profit_margin"] = (
    df["profit"] / df["sales"]
) * 100
```

```python
df.sort_values("profit_margin").head(20)
```

```python
df["delivery_days"] = (
    df["ship_date"] - df["order_date"]
).dt.days
```
